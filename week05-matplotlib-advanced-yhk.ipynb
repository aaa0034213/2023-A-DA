## plot()복습

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False

plt.figure(figsize=(5,3))

x=np.linspace(-2*np.pi,2*np.pi, 200)
plt.plot(x, np.sin(x)*10)
plt.plot(x,x**2)
plt.plot(x,2*x**2-3*x)

plt.grid(True)
plt.title('그래프')
plt.show()
```

## boxplot() 함수

```python
import matplotlib.pyplot as plt
import random

result=[]
for i in range(13):
    result.append(rd.randint(1,1000))
    
print(sorted(result))

plt.boxplot(result,whis=0.5)
plt.show()
```

## plt.subplot()

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))
plt.subplot(2,3,1)
plt.subplot(2,3,5)
plt.subplot(2,3,6)
plt.show()
```

```python
ax1= plt.subplot(2,1,1)
plt.plot(np.random.randint(1,10,10),'yo-')
plt.title('그래프')
plt.ylabel('Damped oscillation')
print(ax1)

ax2=plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)*4,'k.-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

## plt.subplots()

```python
import matplotlib.pyplot as plt

fig, axes= plt.subplots(2,2)

axes[0,0].plot([1,2,3,4])
axes[0,0].set_title("ex 1")

axes[0,1].plot([4,3,2,1])
axes[0,1].set_title("ex 2")

axes[1,0].hist([1,2,2,3,3,3,4,5,6,6],bins=5)
axes[1,0].set_title("ex 3")

axes[1,1].scatter([1,2,3,4],[4,3,2,1])
axes[1,1].set_title("ex 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes= plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title("axes 1")
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title("axes 2")
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title("axes 3")
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title("axes 4")

plt.tight_layout()
plt.show()
```
