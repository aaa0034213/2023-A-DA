```python
# A열 값 B열 값 나눈 결과로 D열 출력
import pandas as pd
import numpy as np

index = pd.date_range('12/1/2024', periods=8)

df = pd.DataFrame(np.random.rand(8,3), index=index , columns=['A','B','C'])
df
```

```python
df['D']= df['A']/df['B']
df
```

```python
df['E'] = np.sum(df, axis=1)   # 같은 행 데이터 값 + 결과 E에 출력 axis=1 중요.
df
```

```python
df['F']= np.ones(8)
df.head(3)
```

```python
# 특정 열 제거
df1= df.drop(['E','F'],axis='columns')  # E,F 칼럼 삭제 후 저장.
df1.head()      # ()지정해주지 않으면 처음 5개만.
```

```python
del df1['D']    # del 로 한 번에 하나열 삭제 가능.
df1.head()
```

```python
df1.columns = ['a','b','c']   # 전체 열 변경
df1 
```

```python
# 특정 열 변경

df1.rename(columns={'a':'2020','b':'2021'},inplace=True)
df1.tail()
```

```python
df1= df1.sub(df1['2020'],axis='index')   # 전체 데이터에서 2020 열 값 빼기
df1.tail()
```

```python
df1=df1.div(df1['c'],axis='index') # 전제 데이터 대해 c 열 값 나누기
df1.tail()
```

```python
df1.to_csv('test.csv')
df1
```
