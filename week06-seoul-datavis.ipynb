```python
# 매년 12/25 평균 최저 최고 기온 데이터 그래프화
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data= csv.reader(f)

header= next(data)

mean=[]
low=[]
high=[]

for row in data:
    if(row[2]!='') and (row[3]!='') and (row[4]!=''):
        if(row[0].split('-')[1]=='12') and (row[0].split('-')[2]=='25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.Figure(figsize=(10,2))
plt.plot(mean,'k')
plt.plot(high,'r')
plt.plot(low,'b')
plt.xlabel('Day (December 25)')
plt.ylabel('temperature')
plt.show()
```

```python
# 평균 기온 변화 그래프
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)

header= next(data)

temp_day=[]
temp_mean=[]
year=[]

for row in data:
    if int(row[0].split('-')[0])>1907:
        if row[2]!='':
            temp_day.append(float(row[2]))
            if(row[0].split('-')[1]=='12') and (row[0].split('-')[2]=='31'):
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day)/ len(temp_day))
                temp_day=[]

f.close()

plt.Figure(figsize=(10,2))
plt.rc('font', family='Malgun Gothic')
plt.title('평군 기온 변화 그래프')
plt.plot(year,temp_mean,color='r', marker='d')
plt.xlabel('Year')
plt.ylabel('AVerage temperature')
plt.show()
```

# 기온 데이터 hist

```python
# 최고 기온 데이터를 hist
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
result= []

for row in data:
    if row[4]!='':
        result.append(float(row[4]))
        
plt.hist(result, bins=100, color='r')
plt.show()
```

```python
# 8월의 최고 기온 데이터 hist
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
aug= []

for row in data:
    month = row[0].split('-')[1]
    if row[4]!='':
        if month == '08':
            aug.append(float(row[4]))
        
plt.hist(aug, bins=100, color='r')
plt.show()
```

```python
# 1월과 8월의 최고 기온 데이터를 hist
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
jan=[]
aug= []


for row in data:
    month = row[0].split('-')[1]
    if row[4]!='':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))
            
plt.hist(jan, bins=100, color='b', label='JAP.')
plt.hist(aug, bins=100, color='r', label='AUG.')
plt.legend()
plt.show()
```

```python
# 최고 기온 데이터를 boxplot으로 시각화

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
result= []

for row in data:
    if row[4]!='':
        result.append(float(row[4]))
        
plt.boxplot(result)
plt.show()
```

```python
# 1월과 8월의 데이터를 boxplot으로 시각화

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[4]!='':
        if month =='01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))
        
plt.boxplot(jan)
plt.boxplot(aug)
plt.show()
```

```python
# 1월과 8월의 데이터를 boxplot으로 시각화 [] , 나눠서 보기

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[4]!='':
        if month =='01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))
        
plt.boxplot([jan,aug])
plt.show()
```

```python
# 1~12까지의 최고 기온 데이터 boxplot

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[4]!='':
        month[int(row[0].split('-')[1])-1].append(float(row[4]))
        
plt.boxplot(month)
plt.show()
```

```python
# 1월 일별 기온 데이터 boxplot

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data= csv.reader(f)
header= next(data)

day= []
for row in range(31):
    day.append([])

for row in data:
    if row[4]!='':
        if row[0].split('-')[1]=='01':
            day[int(row[0].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot')
plt.Figure(figsize=(10,5))
plt.boxplot(day,showfliers=False)
plt.show()
```
