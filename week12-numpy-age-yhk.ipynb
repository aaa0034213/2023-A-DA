```python
import numpy as np   # mask use
a = np.arange(-5,5)
print(a)

mask1= (abs(a)>3)
print(mask1)
print(a[mask1])

mask2 = ((abs(a)%2)==0)
print(mask2)
print(a[mask2])


print(a[mask1+mask2])  # or 연산 
print(a[mask1*mask2])  # and 연산
```

```python
# mask 이용한 버블 차드 그리기 

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

mask1= (abs(x)>50)
mask2= (abs(y)>50)
x = x[mask1+mask2]          # +(or) 인하여 0~49 부분 색x
y = y[mask1+mask2]

plt.scatter(x,y,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# mask 이용한 버블 차드 그리기 

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

mask1= (abs(x)>50)
mask2= (abs(y)>50)
x = x[mask1*mask2]       # *(and) 했을때 그래프 결과 각 영역에만 표시
y = y[mask1*mask2]

plt.scatter(x,y,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('age.csv', encoding='cp949')
data= csv.reader(f)

header = next(data)

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)를 입력해주세요:')
for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)
    if '화촌면' in row[0]:
        homeB = np.array(row[3:],dtype=int)
        
f.close()

plt.rc('font',family='Malgun Gothic')
plt.plot(home, label=name)
plt.plot(homeB, label='화촌면')
plt.title(name+'지역의 인구 구조')
plt.grid(True)
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('age.csv', encoding='cp949')
data= csv.reader(f)

header= next(data)

data=list(data)

min_val=1
result_name=''
result=0

name= input('인구 구조 알고 싶은 지역이름(읍면동)단위를 입력:')


    
for row in data:
    for i in range(102):
        row[i+2]= int(row[i+2].replace(',',""))
    if name in row[0]:
        home = np.array(row[3:],dtype=int)/row[2]
        
for row in data:
    if row[2] ==0:
        continue
    
for row in data:
    if name not in row[0]:
        away = np.array(row[3:],dtype=int) / row[2]
    s= np.sum((home-away**2))
    if s < min_val:
        min_val= s
        result_name = row[0]
        result = away
        
f.close()

print(result_name)
plt.plot(home)
plt.plot(result)
plt.show()
```
